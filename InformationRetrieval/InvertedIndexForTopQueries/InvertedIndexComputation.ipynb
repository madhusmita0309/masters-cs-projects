```python
class invertedIndex(object):
    def __init__(self,docs):
        self.docs = docs
        self.termList = []
        self.docLists = []
        for index,doc in enumerate(docs):
            for term in doc:
                term=term.lower()
                if term.isalpha() and term not in sw:
                    
                    if term in self.termList:
                        i=self.termList.index(term)
                        if index not in self.docLists[i]:
                            self.docLists[i].append(index)
                    else:
                        self.termList.append(term)
                        self.docLists.append([index])
        #print(self.termList)
        #print(self.docLists)
        print('Vocab size', len(self.termList))
        global vocabSize
        vocabSize=len(self.termList)
        #dictsize=0
        #for p,posti in enumerate(self.docLists) :
         #   for z in posti:
          #      dictsize += 1
                
        print('Dictionary size ',len(self.termList))
        p=0
        postingListSize=0
        for p in self.docLists: 
            postingListSize+=len(p)
        print('Posting List Size',postingListSize) 
        dict={}
        
        for p,posti in enumerate(self.docLists) :
            dict.update({self.termList[p]:len(self.docLists[p])})
            #print(len(self.docLists[p]))
        #for (key, value) in dict.items() :
            #print(key , " :: ", value )
        s = [(k, dict[k]) for k in sorted(dict, key=dict.get, reverse=True)]
        
        print("Terms in decreasing order of frequency :")
        for key, value in s:
            print(key," : ",value)
            
    def search(self,term):
           try:
            i=self.termList.index(term)
            return self.docLists[i]
           except:
            return ""
```
